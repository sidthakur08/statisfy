```python
import pandas as pd
import Metrica_IO as mio
import Metrica_Viz as mviz

datadir = '/Users/sidthakur08/ML/Code/Soccer/soccermatics-py/tracking/data'
gameid = 2
```

```python
events = mio.read_event_data(datadir,gameid)
events.head()
```

```python
passes = events[events['Type']=='PASS']
passes['Subtype'].value_counts()
```

```python
passes_home = passes[passes['Team']=='Home'].loc[:,['Type','From','To']]
passes_away = passes[passes['Team']=='Away'].loc[:,['Type','From','To']]
```

```python
home_freq = passes_home.groupby(['From','To']).size().reset_index(name='Time').sort_values('Time',ascending=False).reset_index(drop=True)
home_freq
```

```python
away_freq = passes_away.groupby(['From','To']).size().reset_index(name='Time').sort_values('Time',ascending=False).reset_index(drop=True)
away_freq
```

```python
home_freq.to_csv('home_pass.csv',index=False)
away_freq.to_csv('away_pass.csv',index=False)
```
